Dropping columns that I don't need

```python

movies = movies.drop(['budget', 'revenue', 'runtime', 'status', 'poster_path', 'backdrop_path', 'recommendations', 'production_companies'], axis = 1)
movies
```

Dropping na

```python
movies = movies.dropna(how = 'all')
movies
```

```python
movies.fillna("None", inplace=True)
```

Checking duplicates

```python
movies = movies.drop_duplicates()
movies
```

## Release Date Modification (in process)

Modifying release date so I only get the year. This way I can standarize data.

```python
movies['release_year'] = movies['release_date'].str[:4]
movies
```

```python
movies = movies.drop(['release_date'], axis=1)
```

## Genre modification (in process)

Seeing that data is not standarized by checking value_counts

```python
movies['genres'].value_counts()
```

Checking how many unique genres are there

```python
new_dict_ = {}

for i in list(movies["genres"]):
    if type(i) != float:
        genres = i.split("-")
        for j in genres:
            new_dict_[j] = 0
    else:
        pass

new_dict_
```

Checking the numerical value of each genre

```python
genre_count = {}

for i in list(movies["genres"]):
    if type(i) != float:
        genres = i.split("-")
        for j in genres:
            if j in genre_count:
                genre_count[j] += 1
            else:
                genre_count[j] = 1
    else:
        pass

new_dict_
```

Graphical Representation

```python
sorted_genres = sorted(new_dict_.keys(), key=lambda x: new_dict_[x], reverse=True)
sorted_counts = [new_dict_[genre] for genre in sorted_genres]

plt.bar(sorted_genres, sorted_counts)
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Genre Counts in Movies (Descending Order)')
plt.xticks(rotation=45, ha='right')  

plt.show()
```
